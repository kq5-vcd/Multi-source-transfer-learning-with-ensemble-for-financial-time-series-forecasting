```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Explore the dataset used

```python
teacher_path = '../stock_data/data_for_teacher'
student_path = '../stock_data/data_for_student'
```

```python
print("Student")

student_data = {}
for corp in os.listdir(student_path):
    df = pd.read_csv(student_path + "/" + corp)
    print(corp[:-4], len(df))
    
    student_data[corp[:-4]] = df
    
print("Teacher")

teacher_data = {}
for corp in os.listdir(teacher_path):
    df = pd.read_csv(teacher_path + "/" + corp)
    print(corp[:-4], len(df))
    
    teacher_data[corp[:-4]] = df
```

## Visualization

```python
for corp, df in student_data.items():
    plt.plot(df.close)
    plt.title(corp)
    plt.show()
    plt.clf()
```

```python
for corp, df in teacher_data.items():
    plt.plot(df.close)
    plt.title(corp)
    plt.show()
    plt.clf()
```
