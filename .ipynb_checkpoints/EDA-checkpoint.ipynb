```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Explore the dataset used

```python
G1 = ["BOC", "BOCHK", "BOCOM", "CCB", "HSB", "HSBC", "ICBC"]
G1
```

```python
data = {}

for corp in G1:
    df = pd.read_csv("Data/" + corp + ".csv")
    
    print(corp)
    print(df.describe())
    
    data[corp] = df[["Date", "Close"]]
```

## Visualization

```python
for corp, df in data.items():
    plt.plot(df.Close)
    plt.title(corp)
    plt.show()
    plt.clf()
```
